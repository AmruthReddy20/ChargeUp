```python
def input_to_output(input_EVPR, input_HCR):
    # finding the index of the number that is closest to the input demand_ratio
    demand_ratio_input =  input_EVPR*(0.1*1+0.9*(1-input_HCR)/5)
    list_index = int(np.round((demand_ratio_input-0.00118)/((0.038-0.00118)/29)))
    demand_ratio = list_demand_ratio[list_index]
    # return the map html file
    map_html = mypath + 'graphs/gmplot/map_demand_ratio_'+str(demand_ratio)+'.html'
    # return the table of the optimal locations of the parking stations
    df_opt_chg_lc = pd.read_excel(mypath +'data/processed/chg_stn_location_demand_ratio'+str(demand_ratio)+'.xlsx')
    html_opt_chg_lc = df_opt_chg_lc.to_html()
    # return table: how many more charging stations do we need to build in each ward
    # will work on this later! 
    return map_html, df_opt_chg_lc
```
